```python
from abc import ABC, abstractmethod

# Sistema de animales

class Animal(ABC):
    def __init__(self, nombre, especie, edad):
        self.nombre = nombre
        self.especie = especie
        self.__edad = edad          # Encapsulamiento
        self.__energia = 100        # Encapsulamiento

    # Abstracción
    @abstractmethod
    def tipo_animal(self):
        pass

    # Encapsulamiento: getters
    def obtener_edad(self):
        return self.__edad

    def obtener_energia(self):
        return self.__energia

    # Polimorfismo: método que será redefinido
    def hacer_sonido(self):
        print(f"{self.nombre} hace un sonido genérico.")

    def comer(self, alimento):
        self.__energia += 20
        print(f"{self.nombre} come {alimento}. Energía: {self.__energia}")

    def mostrar_info(self):
        print(f"Nombre: {self.nombre} | Especie: {self.especie} | Edad: {self.__edad} años | Energía: {self.__energia}")


# Herencia
class Perro(Animal):
    def __init__(self, nombre, edad, raza):
        super().__init__(nombre, "Canis lupus familiaris", edad)
        self.raza = raza

    def tipo_animal(self):
        return "Mamífero doméstico"

    # Polimorfismo
    def hacer_sonido(self):
        print(f"{self.nombre} ({self.raza}) dice: ¡Guau guau!")

    def buscar_pelota(self):
        print(f"{self.nombre} corre a buscar la pelota")


# Herencia
class Pajaro(Animal):
    def __init__(self, nombre, edad, puede_volar):
        super().__init__(nombre, "Ave", edad)
        self.puede_volar = puede_volar

    def tipo_animal(self):
        return "Ave"

    # Polimorfismo
    def hacer_sonido(self):
        print(f"{self.nombre} dice: ¡Pío pío!")

    def volar(self):
        if self.puede_volar:
            print(f"{self.nombre} vuela alto en el cielo")
        else:
            print(f"{self.nombre} no puede volar.")


print("=" * 50)
print("        OBJETO 1: SISTEMA DE ANIMALES")
print("=" * 50)

perro = Perro("Kora", 6, "Labrador")
perro.mostrar_info()
perro.hacer_sonido()
perro.comer("croquetas")
perro.buscar_pelota()
print(f"Tipo: {perro.tipo_animal()}")

print()
pajaro = Pajaro("Pico", 1, True)
pajaro.mostrar_info()
pajaro.hacer_sonido()
pajaro.volar()
pajaro.comer("semillas")
print(f"Tipo: {pajaro.tipo_animal()}")


# Sistema de productos de tienda

class Producto(ABC):
    def __init__(self, nombre, codigo, precio):
        self.nombre = nombre
        self.codigo = codigo
        self.__precio = precio      # Encapsulamiento
        self.__stock = 0            # Encapsulamiento

    # Abstracción
    @abstractmethod
    def categoria(self):
        pass

    # Encapsulamiento: getters y setters
    def obtener_precio(self):
        return self.__precio

    def obtener_stock(self):
        return self.__stock

    def actualizar_precio(self, nuevo_precio):
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
            print(f"Precio de '{self.nombre}' actualizado a: ${self.__precio:,.0f}")
        else:
            print("El precio debe ser mayor a 0.")

    def agregar_stock(self, cantidad):
        if cantidad > 0:
            self.__stock += cantidad
            print(f"Stock de '{self.nombre}' actualizado. Stock actual: {self.__stock} unidades")

    # Polimorfismo
    def aplicar_descuento(self):
        descuento = self.__precio * 0.05
        print(f"Descuento estándar en '{self.nombre}': ${descuento:,.0f}")
        return descuento

    def mostrar_info(self):
        print(f"[{self.codigo}] {self.nombre} | Categoría: {self.categoria()} | Precio: ${self.__precio:,.0f} | Stock: {self.__stock}")


# Herencia
class ProductoElectronico(Producto):
    def __init__(self, nombre, codigo, precio, garantia_meses):
        super().__init__(nombre, codigo, precio)
        self.garantia_meses = garantia_meses

    def categoria(self):
        return "Electrónico"

    # Polimorfismo: descuento mayor para electrónicos
    def aplicar_descuento(self):
        descuento = self.obtener_precio() * 0.15
        print(f"Descuento electrónico en '{self.nombre}' ({self.garantia_meses} meses garantía): ${descuento:,.0f}")
        return descuento


# Herencia
class ProductoAlimenticio(Producto):
    def __init__(self, nombre, codigo, precio, fecha_vencimiento):
        super().__init__(nombre, codigo, precio)
        self.fecha_vencimiento = fecha_vencimiento

    def categoria(self):
        return "Alimento"

    # Polimorfismo: descuento por vencimiento próximo
    def aplicar_descuento(self):
        descuento = self.obtener_precio() * 0.30
        print(f"Descuento por vencimiento '{self.nombre}' (vence: {self.fecha_vencimiento}): ${descuento:,.0f}")
        return descuento


print("\n" + "=" * 50)
print("     OBJETO 2: SISTEMA DE PRODUCTOS")
print("=" * 50)

laptop = ProductoElectronico("Laptop HP", "ELEC-001", 3500000, 12)
laptop.agregar_stock(10)
laptop.mostrar_info()
laptop.aplicar_descuento()
laptop.actualizar_precio(3200000)

print()
leche = ProductoAlimenticio("Leche Entera 1L", "ALIM-023", 4200, "2025-05-01")
leche.agregar_stock(50)
leche.mostrar_info()
leche.aplicar_descuento()
leche.actualizar_precio(-100)   # Prueba de validación

# Polimorfismo en acción
print("\n--- Polimorfismo: descuentos de todos los productos ---")
productos = [laptop, leche, ProductoElectronico("Audífonos Sony", "ELEC-002", 250000, 6)]
for p in productos:
    p.aplicar_descuento()


# Sistema de cursos académicos

class Curso(ABC):
    def __init__(self, nombre, codigo, duracion_horas):
        self.nombre = nombre
        self.codigo = codigo
        self.__duracion_horas = duracion_horas   # Encapsulamiento
        self.__estudiantes = []                  # Encapsulamiento

    # Abstracción
    @abstractmethod
    def modalidad(self):
        pass

    # Encapsulamiento: getters
    def obtener_duracion(self):
        return self.__duracion_horas

    def obtener_num_estudiantes(self):
        return len(self.__estudiantes)

    def inscribir_estudiante(self, nombre_estudiante):
        self.__estudiantes.append(nombre_estudiante)
        print(f"'{nombre_estudiante}' inscrito en '{self.nombre}'. Total: {len(self.__estudiantes)} estudiantes")

    # Polimorfismo
    def calcular_costo(self):
        costo = self.__duracion_horas * 10000
        print(f"Costo base de '{self.nombre}': ${costo:,.0f}")
        return costo

    def mostrar_info(self):
        print(f"[{self.codigo}] {self.nombre} | Modalidad: {self.modalidad()} | Duración: {self.__duracion_horas}h | Estudiantes: {len(self.__estudiantes)}")


# Herencia
class CursoPresencial(Curso):
    def __init__(self, nombre, codigo, duracion_horas, salon):
        super().__init__(nombre, codigo, duracion_horas)
        self.salon = salon

    def modalidad(self):
        return "Presencial"

    # Polimorfismo: costo mayor por uso de instalaciones
    def calcular_costo(self):
        costo = self.obtener_duracion() * 15000
        print(f"Costo presencial '{self.nombre}' (salón {self.salon}): ${costo:,.0f}")
        return costo


# Herencia
class CursoVirtual(Curso):
    def __init__(self, nombre, codigo, duracion_horas, plataforma):
        super().__init__(nombre, codigo, duracion_horas)
        self.plataforma = plataforma

    def modalidad(self):
        return "Virtual"

    # Polimorfismo: costo menor por ser en línea
    def calcular_costo(self):
        costo = self.obtener_duracion() * 7000
        print(f"Costo virtual '{self.nombre}' (plataforma: {self.plataforma}): ${costo:,.0f}")
        return costo


print("\n" + "=" * 50)
print("    OBJETO 3: SISTEMA DE CURSOS")
print("=" * 50)

curso_presencial = CursoPresencial("Matemáticas Avanzadas", "MAT-301", 48, "Aula 204")
curso_presencial.inscribir_estudiante("Natalia Camargo")
curso_presencial.inscribir_estudiante("Heidy Velez")
curso_presencial.mostrar_info()
curso_presencial.calcular_costo()

print()
curso_virtual = CursoVirtual("Python para Data Science", "PROG-102", 30, "Moodle")
curso_virtual.inscribir_estudiante("Sofía Ramírez")
curso_virtual.mostrar_info()
curso_virtual.calcular_costo()

# Polimorfismo en acción
print("\n--- Polimorfismo: costos de todos los cursos ---")
cursos = [curso_presencial, curso_virtual, CursoVirtual("Inglés B2", "IDIO-201", 60, "Teams")]
for c in cursos:
    c.calcular_costo()
```
